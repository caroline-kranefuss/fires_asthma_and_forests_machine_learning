```python
import pandas as pd
```

```python
# Load the data
two = pd.read_csv('asthma_2022.txt')
three = pd.read_csv('asthma_2023.txt')
#five = 

# Look at a sample df 
two.head()
```

```python
# Rename columns - doing individually due to year in columns

two = two.rename(columns={'2022 National Rankings': '2022_Rank',
                 'Metropolitan Area': 'Metropolitan_Area',
                 'Total Score': 'Score'})

three = three.rename(columns={'2023 National Rankings': '2023_Rank',
                 'Metropolitan Area': 'Metropolitan_Area',
                 'Total Score': 'Score'})

"""
five = five.rename(columns={'2025 National Rankings': '2025_Rank',
                 'Metropolitan Area': 'Metropolitan_Area',
                 'Total Score': 'Score'})
"""
```

```python
# Make a state column - can be written as a function to be applied to each dataframe

def state(df):
    df['State'] = df['Metropolitan_Area'].str[-2:]
    return df

state(two)
state(three)
# state(five)
```

```python
# Make each county entry a list

def counties_list(df):
    df['Counties'] =  df['Counties'].astype(str).str.split(', ')
    return df

counties_list(two)
counties_list(three)
# counties_list(five)
```

```python
# Split up each df so that each county has its own row. There may be some duplicates... take the average if this is the case

def split(df):
    for i in range(len(df)):
        for j in range(len(df.loc[i, 'Counties'])):
                df['County'] = df.loc[i, 'Counties'][j]

def split_h(df):
     for i in range(1):
          for j in range(4):
               df['County'] = df.loc[i, 'Counties'][j]

split(two)
split_h(three)
# split(five)
three

```
