```python
import pandas as pd
```

```python
# Load the data
two = pd.read_csv('asthma_2022.txt')
three = pd.read_csv('asthma_2023.txt')
#five = 

# Look at a sample df 
two.head()
```

```python
# Rename columns - doing individually due to year in columns

two = two.rename(columns={'2022 National Rankings': 'Rank',
                 'Metropolitan Area': 'Metropolitan_Area',
                 'Total Score': 'Score'})

three = three.rename(columns={'2023 National Rankings': 'Rank',
                 'Metropolitan Area': 'Metropolitan_Area',
                 'Total Score': 'Score'})

"""
five = five.rename(columns={'2025 National Rankings': 'Rank',
                 'Metropolitan Area': 'Metropolitan_Area',
                 'Total Score': 'Score'})
"""
```

```python
# Make a state column - can be written as a function to be applied to each dataframe

def state(df):
    df['State'] = df['Metropolitan_Area'].str[-2:]
    return df

state(two)
state(three)
# state(five)
```

```python
# Make each county entry a list

def counties_list(df):
    df['Counties'] =  df['Counties'].astype(str).str.split(', ')
    return df

counties_list(two)
counties_list(three)
# counties_list(five)
```

```python
# Split up each df so that each county has its own row. There may be some duplicates... take the average if this is the case

def split(df):
    # Add a column with empty strings for now
    df['County'] = ''
    # Store the length of the original dataframe
    orig_len = len(df)
    # For every row in the dataframe
    for i in range(len(df)):
        for j in range(len(df.loc[i, 'Counties'])):
            # Create a new row at the end
            df.loc[len(df)] = [df.loc[i, 'Rank'], 
                            df.loc[i, 'Metropolitan_Area'],
                            df.loc[i, 'Score'],
                            df.loc[i, 'Counties'], 
                            df.loc[i, 'State'], 
                            #For the County column, take the jth item in the list
                            df.loc[i,'Counties'][j]
                            ]
    # Remove the initial rows that don't have a county
    df.drop(labels=range(orig_len), axis=0, inplace=True)
    return df

split(two)
split(three)
# split(five)

```

```python
# Make a State_County column
    # Where County's state differs from State's state, use County's state 

def state_county(df):
    df['County'] = df['County'].apply(lambda x: x.replace(' ', '_'))   
    # If the states match
    df_match = df[df['State'] == df['County'].str[-2:]]
    # If they don't match
    df_not = df[df['State'] != df['County'].str[-2:]]
    #if df['State'] == df['County'].str[-2:]:
    #    df['State_County'] = df['State'] + '_' + df['County'] 
    return df_not

state_county(two)
state_county(three)
# state_county(five)
```
